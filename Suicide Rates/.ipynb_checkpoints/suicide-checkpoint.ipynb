```python
# suicidesByCnY = pd.DataFrame(suicide, columns=['country', 'year', 'suicides_no'])

suicidesByCnY = suicide[['country', 'year', 'suicides_no']].groupby(['country','year'], as_index=False).suicides_no.sum()



# duplicated = suicidesByCnY[suicidesByCnY.duplicated(['country'])]
# suicidesByCnY = suicidesByCnY.groupby(["country", "year"]).agg({"suicides_no": ["sum"]})

# .suicides_no.sum()).to_frame()
# duplicated = (duplicated.groupby(["year"])).suicides_no.sum().to_frame()
# duplicated.columns = duplicated.columns.to_series().apply(lambda x: x.strip())
suicidesByCnY
```

```python
# barGraph = sns.FacetGrid(suicidesByCnY, row="country")
barGraph = sns.lmplot(x='year', y='suicides_no', data=suicidesByCnY, row='country', sharex=False, sharey=False)
barGraph.set(ylim=(0, None))
# barGraph.map(plt.hist, "year")
# barGraph.map_dataframe(sns.lineplot, x="year", y="suicides_no")
barGraph.set_axis_labels("year", "number of suicides")
```

With the regression plot of the total numbe of suicides a year, we can pinpoint spikes within the data points but also, determine the mental wellbeing of the citizens of a country to a certain degree. It could offer insight to try to look after the lives of citizens and their well being.
